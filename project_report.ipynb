```r
#changing y to be subscribed to be more readable
data <- mutate(data, subscribed = as_factor(y))
data <- mutate(data, euribor3m = as.numeric(euribor3m))
data <- select(data, "age", 'duration', 'emp.var.rate', 'cons.price.idx', 'cons.conf.idx', 'euribor3m', 'nr.employed', 'subscribed')
```

```r
#splitting training data and testing data to be 75/25 training and testing split, respectively
bank_split <- initial_split(data, prop = 0.75, strata = subscribed)
bank_train <- training(bank_split)
bank_test <- testing(bank_split)
head(bank_train, 15)
```

```r
#standardizing and upsampling training data for the graph
bank_recipe_graph <- recipe(subscribed ~ duration + age , data = bank_train) %>%
    step_scale(all_predictors()) %>%
    step_center(all_predictors()) %>%
    step_upsample(subscribed, over_ratio=1, skip=FALSE) %>%
    prep()

scaled_upsampled_bank <- bake(bank_recipe_graph, bank_train)
```

```r
#showing counts between yes and no
group_by(bank_train, subscribed) %>% summarize(count=n())
```

```r
#showing counts between yes and no after upsampling
group_by(scaled_upsampled_bank, subscribed) %>% summarize(count=n())
```

```r
#plotting the data
ggplot(scaled_upsampled_bank, aes(x=age, y=duration, color = subscribed)) +
    geom_point(alpha=0.6) + 
    labs(x="Age (Scaled)", y= "Duration of call in seconds (Scaled)", color = "Subscribed") + 
    ggtitle("Age vs Duration of call") + 
    theme(text = element_text(size=18), plot.title = element_text(hjust = 0.5))
```

##### <span style='color:gray'>Figure 6: Table count of people who have subscribed to a term deposit in the training data after upsampling</span>
---

```r
#upsampling and centering for cross validation
bank_recipe <- recipe(subscribed ~ duration + age , data = bank_train) %>%
    step_upsample(subscribed, over_ratio=1, skip=FALSE) %>%
    prep()

bank_up_data <- bake(bank_recipe, bank_train)

bank_up_recipe <- recipe(subscribed ~ duration + age , data = bank_up_data) %>%
    step_scale(all_predictors()) %>%
    step_center(all_predictors())

#creating 5-fold cross validation
#splitting training data into 5 folds
bank_vfold <- vfold_cv(bank_up_data, v=5, strata = subscribed)
```

```r
#to tune for best k-neighbors
knn_spec <- nearest_neighbor(weight_func = "rectangular", neighbors = tune()) %>%
    set_engine("kknn") %>%
    set_mode("classification")

#finding best number of neighbors between 1 to 18
grid_val <- tibble(neighbors = seq(1,18))
knn_results <- workflow() %>%
    add_recipe(bank_up_recipe) %>%
    add_model(knn_spec) %>%
    tune_grid(resamples= bank_vfold, grid= grid_val) %>%
    collect_metrics()
```

```r
#displaying the best k neighbor(s)
accuracies <- knn_results %>%
    filter(.metric == "accuracy")
filter(accuracies, mean==max(mean))
best <- filter(accuracies, mean==max(mean)) %>%
    select(neighbors)
```

##### <span style='color:gray'>Figure 7: Table showing the best number of neighbors </span>
---

```r
#accuracy plot
accuracy_vs_k <- ggplot(accuracies, aes(x=neighbors, y=mean)) + 
    geom_point() +
    geom_line() +
    labs(x="Neighbors", y="Accuracy estimate")

accuracy_vs_k
```

##### <span style='color:gray'>Figure 8: Plot of neighbors versus accuracy estimate </span>
---

```r
#new nearest neighbor spec which uses the neighbor with greatest accuracy
knn_spec <- nearest_neighbor(weight_func = "rectangular", neighbors = best$neighbors[1]) %>%
    set_engine("kknn") %>%
    set_mode("classification")
bank_fit <- workflow() %>%
    add_recipe(bank_up_recipe) %>%
    add_model(knn_spec) %>%
    fit(data= bank_train)
```

```r
#calculating the accuracy of the best neighbor
bank_predictions <- predict(bank_fit, bank_test) %>%
    bind_cols(bank_test)%>%
    metrics(truth = subscribed, estimate = .pred_class) %>%
    filter(.metric == "accuracy") %>%
    select(-.estimator)
bank_predictions
```

We will be using bank_addition.csv and **keeping** the columns listed below. The following  variables will be kept because they are numerical and thus can be used to make a predictive model. We also believe that some of these variables will correlate with the target variables (whether a given person subscribes to a term deposit) and therefore will influence the predictor’s accuracy.

* `pdays`: number of days since the client was last contacted from a previous call
* `previous`: number of times this client was contacted before this campaign
